```python
df = df[df['Unnamed: 32'].isnull()]
df.info()
```

```python
df = df[df['Unnamed: 31'].isnull()]
df.info()
```

```python
df = df.rename(columns={'Unnamed: 30': 'state'})

df.info()
```

```python
df_clean = df[['title', 'totalyearlycompensation', 'Race', 'yearsofexperience','gender','Education' ,'state']]
df_clean.head()
```

```python
df_clean.head()
df_clean.info()
```

```python
df_clean = df_clean.dropna(subset=['Race', 'gender', 'Education'])
df_clean.info()
```

```python
Q1 = df_clean['totalyearlycompensation'].quantile(0.25)
Q3 = df_clean['totalyearlycompensation'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_clean[(df_clean['totalyearlycompensation'] < lower_bound) | (df_clean['totalyearlycompensation'] > upper_bound)]
outliers
```

```python
no_outliers = df_clean.drop(df_clean[(df_clean['totalyearlycompensation'] < lower_bound) | (df_clean['totalyearlycompensation'] > upper_bound)].index)
no_outliers.info()
```

```python
no_outliers.head()
```

```python
min_value = no_outliers['totalyearlycompensation'].min()
max_value = no_outliers['totalyearlycompensation'].max()

print("Minimum value:", min_value)
print("Maximum value:", max_value)
```

```python
# Define the bucket intervals and labels
start_value = 10000
increment = 20000
end_value = 450000

bucket_intervals = list(range(start_value, end_value + increment, increment))
bucket_labels = [f'${bucket_intervals[i]//1000}k-${(bucket_intervals[i+1])//1000}k' for i in range(len(bucket_intervals)-1)]

# Add the bucketed 'totalyearlycompensation' column as 'comp_bins'
new_df['comp_bins'] = pd.cut(no_outliers['totalyearlycompensation'], bins=bucket_intervals, labels=bucket_labels, right=False)

# Print the updated DataFrame
print(new_df.head())
```

```python
new_df.head()

```

```python
new_df['comp_bins'].unique()
```

```python
new_df.info()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = new_df.drop(columns=['comp_bins'])  # Independent variables
y = new_df['comp_bins']  # Dependent variable


```

```python
X_encoded = pd.get_dummies(X)
X_encoded.head()
X_encoded.info()

```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.1, random_state=42)
```

```python
# Initialize the Decision Tree model
model = DecisionTreeClassifier()

# Fit the model on the training data
model.fit(X_train, y_train)



y_pred = model.predict(X_test)
```

```python
# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```
